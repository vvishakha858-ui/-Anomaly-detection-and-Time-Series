# 1-  What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.


>### Anomaly Detection

**Anomaly Detection** is the process of identifying data points, patterns, or observations that **deviate significantly from normal behavior** in a dataset.
These unusual patterns are called **anomalies** or **outliers** and often indicate **errors, fraud, faults, or rare events**.

It is widely used in **fraud detection, network security, healthcare monitoring, manufacturing, and finance**.


## Types of Anomalies

### 1Ô∏è‚É£ Point Anomalies

A **single data point** that is significantly different from the rest of the data.

**Example:**

* Credit card transactions where most amounts are between ‚Çπ500‚Äì‚Çπ5,000, but one transaction is **‚Çπ2,00,000**
* A temperature sensor usually reports 20‚Äì25¬∞C, but suddenly shows **85¬∞C**

**Key idea:**
‚û° One data point is abnormal **by itself**


### 2Ô∏è‚É£ Contextual Anomalies

A data point is anomalous **only in a specific context** (time, location, condition).

**Example:**

* 30¬∞C temperature is normal in **summer**, but anomalous in **winter**
* High electricity usage is normal during **daytime**, but anomalous at **midnight**
* High website traffic during a sale is normal, but abnormal on a regular weekday

**Key idea:**
‚û° Normal in one context, **abnormal in another**


### 3Ô∏è‚É£ Collective Anomalies

A **group of data points together** forms an anomaly, even if individual points appear normal.

**Example:**

* A sequence of small bank transactions repeatedly occurring within seconds (possible fraud)
* Network packets individually look normal, but a sudden burst indicates a **DDoS attack**
* Gradual but continuous increase in machine vibration before failure

**Key idea:**
‚û° The **pattern or sequence** is anomalous, not individual points



# 2-Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

>### Comparison of Isolation Forest, DBSCAN, and Local Outlier Factor (LOF)

Anomaly detection algorithms differ in how they define and identify **outliers**. Below is a clear comparison of **Isolation Forest**, **DBSCAN**, and **Local Outlier Factor (LOF)** based on their **approach** and **suitable use cases**.


## 1Ô∏è‚É£ Isolation Forest

###  Approach

* Based on the idea that **anomalies are easier to isolate** than normal points.
* Uses **random feature selection and random splits** to build isolation trees.
* Anomalies require **fewer splits** to be isolated.

###  Strengths

* Works well on **high-dimensional data**
* **Fast and scalable** for large datasets
* Does **not rely on distance or density**

###  Suitable Use Cases

* Fraud detection (credit card, insurance)
* Intrusion detection
* Large transactional datasets
* Log anomaly detection



## 2Ô∏è‚É£ DBSCAN

###  Approach

* **Density-based** algorithm
* Groups points into clusters based on **density**
* Points in low-density regions are labeled as **noise (anomalies)**

###  Strengths

* Automatically finds the number of clusters
* Detects **arbitrary-shaped clusters**
* Explicitly identifies **noise/outliers**

###  Suitable Use Cases

* Spatial / geolocation anomaly detection
* Customer behavior clustering with noise
* Sensor data with clear dense regions



## 3Ô∏è‚É£ Local Outlier Factor (LOF)

###  Approach

* Measures **local density deviation**
* Compares a point‚Äôs density with its neighbors
* Points with **significantly lower density** than neighbors are anomalies

###  Strengths

* Excellent at detecting **local anomalies**
* Handles datasets with **varying density**

###  Suitable Use Cases

* Network traffic monitoring
* Detecting abnormal users in behavioral data
* Time-series subsequence anomaly detection







# 3- What are the key components of a Time Series? Explain each with one example.



>### Key Components of a Time Series

A **time series** is a sequence of data points recorded **over time at regular intervals**.
Understanding its components helps in **analysis, forecasting, and anomaly detection**.



## 1Ô∏è‚É£ Trend

**Definition:**
The **long-term upward or downward movement** in the data over time.

**Example:**

* A steady increase in **online sales revenue** over several years due to business growth.

 *Interpretation:* Shows overall direction of the data.


## 2Ô∏è‚É£ Seasonality

**Definition:**
Regular, **repeating patterns** that occur at fixed intervals (daily, weekly, monthly, yearly).

**Example:**

* Higher **e-commerce sales during festive seasons** like Diwali every year.
* Increased electricity consumption during **summer months**.

 *Interpretation:* Patterns that repeat predictably.


## 3Ô∏è‚É£ Cyclical Component

**Definition:**
Fluctuations that occur over **irregular, longer periods** and are often influenced by external factors.

**Example:**

* Business sales rising during economic booms and falling during recessions.

 *Interpretation:* No fixed period; related to economic or business cycles.


## 4Ô∏è‚É£ Irregular / Random (Noise)

**Definition:**
Unpredictable, **random variations** that do not follow a pattern.

**Example:**

* Sudden drop in website traffic due to a **server outage**.
* Unexpected spike in stock price due to breaking news.

 *Interpretation:* Residual variation after removing trend, seasonality, and cycles.





# 4- Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?



>### What is Stationarity in Time Series?

A **time series is called stationary** if its **statistical properties do not change over time**.

#### For a stationary series:

* **Mean** remains constant
* **Variance** remains constant
* **Autocovariance** depends only on the lag, not on time

 Stationarity is important because **most time-series models (ARMA, ARIMA)** assume the data is stationary.


## How to Test Stationarity

### 1Ô∏è‚É£ Visual Inspection

* **Time plot** ‚Üí No clear trend or seasonality
* **Rolling mean & variance** ‚Üí Should be constant over time


### 2Ô∏è‚É£ Augmented Dickey‚ÄìFuller (ADF) Test

* **Null Hypothesis (H‚ÇÄ):** Time series is **non-stationary**
* **Alternative Hypothesis (H‚ÇÅ):** Time series is **stationary**

**Decision Rule:**

* If **p-value < 0.05** ‚Üí Reject H‚ÇÄ ‚Üí Series is stationary
* If **p-value ‚â• 0.05** ‚Üí Non-stationary


### 3Ô∏è‚É£ KPSS Test (Complementary)

* **Null Hypothesis:** Series is **stationary**
* Useful to confirm ADF results


## How to Transform a Non-Stationary Series into a Stationary One

### 1Ô∏è‚É£ Differencing

Removes trend by subtracting previous values.

* **First-order differencing:**
  ( y'*t = y_t - y*{t-1} )

**Example:**
Stock prices ‚Üí stock returns



### 2Ô∏è‚É£ Seasonal Differencing

Removes seasonal effects.

* ( y'*t = y_t - y*{t-s} )
  (where *s* = seasonal period)

**Example:**
Monthly sales data ‚Üí subtract sales from the same month last year



### 3Ô∏è‚É£ Log / Power Transformations

Stabilize **variance**.

* Log transform
* Square root transform
* Box-Cox transform

**Example:**
Sales data with increasing variance over time


### 4Ô∏è‚É£ Detrending

Remove trend using:

* Regression
* Moving average subtraction



## Example Workflow

1. Plot the time series
2. Apply **ADF test**
3. If non-stationary ‚Üí apply **log transform or differencing**
4. Re-test using ADF
5. Repeat until stationarity is achieved






# 5- Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.



### Differentiation Between AR, MA, ARIMA, SARIMA, and SARIMAX Models

These models are widely used for **time series forecasting**, each extending the previous one to handle more complex patterns.


## 1Ô∏è‚É£ AR (AutoRegressive) Model

### üîπ Structure

* Uses **past values** of the series to predict current value.
* Denoted as **AR(p)**, where *p* is the number of lag observations.

[
y_t = c + \phi_1 y_{t-1} + \phi_2 y_{t-2} + \dots + \phi_p y_{t-p} + \varepsilon_t
]

###  Application

* When the series is **stationary**
* Short-term forecasting

**Example:**
Predicting tomorrow‚Äôs temperature using the last few days‚Äô temperatures.

---

## 2Ô∏è‚É£ MA (Moving Average) Model

### üîπ Structure

* Uses **past forecast errors (residuals)**.
* Denoted as **MA(q)**, where *q* is the number of lagged errors.

[
y_t = c + \varepsilon_t + \theta_1 \varepsilon_{t-1} + \dots + \theta_q \varepsilon_{t-q}
]

###  Application

* Stationary series with **shock effects**

**Example:**
Modeling sudden demand changes due to promotions.


## 3Ô∏è‚É£ ARIMA (AutoRegressive Integrated Moving Average)

### üîπ Structure

* Combines **AR(p)** + **Differencing(d)** + **MA(q)**.
* Written as **ARIMA(p, d, q)**.

###  Application

* Non-stationary data that becomes stationary after differencing
* No seasonality

**Example:**
Stock price forecasting after differencing to remove trend.


## 4Ô∏è‚É£ SARIMA (Seasonal ARIMA)

### üîπ Structure

* Extends ARIMA by adding **seasonal components**.
* Written as **SARIMA(p, d, q)(P, D, Q, s)**
  where *s* = seasonal period.

###  Application

* Time series with **seasonality**

**Example:**
Monthly sales with yearly seasonality (*s = 12*).


## 5Ô∏è‚É£ SARIMAX (SARIMA with Exogenous Variables)

### üîπ Structure

* SARIMA + **external (exogenous) variables**.
* Uses additional predictors besides past values.

###  Application

* When external factors influence the series

**Example:**
Sales forecasting using:

* Historical sales
* Advertising spend
* Discounts

# 7- Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.



```python



import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

np.random.seed(42)

normal_trips = pd.DataFrame({
    "trip_distance": np.random.normal(loc=5, scale=2, size=450),
    "fare_amount": np.random.normal(loc=15, scale=5, size=450)
})

anomalies = pd.DataFrame({
    "trip_distance": np.random.uniform(20, 40, size=20),
    "fare_amount": np.random.uniform(80, 150, size=20)
})

df = pd.concat([normal_trips, anomalies], ignore_index=True)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.04,
    random_state=42
)

df["anomaly"] = iso_forest.fit_predict(X_scaled)

normal = df[df["anomaly"] == 1]
outliers = df[df["anomaly"] == -1]

plt.figure(figsize=(8, 6))
plt.scatter(normal["trip_distance"], normal["fare_amount"],
            c="blue", s=30, label="Normal Trips")
plt.scatter(outliers["trip_distance"], outliers["fare_amount"],
            c="red", s=60, marker="x", label="Anomalies")

plt.xlabel("Trip Distance (km)")
plt.ylabel("Fare Amount ($)")
plt.title("Isolation Forest ‚Äì Anomaly Detection on NYC Taxi Fare Data")
plt.legend()
plt.show()
```

# 8-Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.



```python



import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

data = {
    "Month": pd.date_range(start="1949-01-01", periods=144, freq="M"),
    "Passengers": [
        112,118,132,129,121,135,148,148,136,119,104,118,
        115,126,141,135,125,149,170,170,158,133,114,140,
        145,150,178,163,172,178,199,199,184,162,146,166,
        171,180,193,181,183,218,230,242,209,191,172,194,
        196,196,236,235,229,243,264,272,237,211,180,201,
        204,188,235,227,234,264,302,293,259,229,203,229,
        242,233,267,269,270,315,364,347,312,274,237,278,
        284,277,317,313,318,374,413,405,355,306,271,306,
        315,301,356,348,355,422,465,467,404,347,305,336,
        340,318,362,348,363,435,491,505,404,359,310,337,
        360,342,406,396,420,472,548,559,463,407,362,405,
        417,391,419,461,472,535,622,606,508,461,390,432
    ]
}

df = pd.DataFrame(data)
df.set_index("Month", inplace=True)

model = SARIMAX(
    df["Passengers"],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

results = model.fit()

forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

plt.figure(figsize=(10, 5))
plt.plot(df["Passengers"], label="Observed")
plt.plot(forecast_mean, label="Forecast", color="red")
plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    color="pink",
    alpha=0.3,
    label="Confidence Interval"
)

plt.title("SARIMA Forecast ‚Äì AirPassengers (Next 12 Months)")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

# 9- Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.

```python



import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.preprocessing import StandardScaler

np.random.seed(42)


X_normal = np.random.normal(loc=0, scale=1, size=(300, 2))

X_anomalies = np.random.uniform(low=-6, high=6, size=(20, 2))

X = np.vstack([X_normal, X_anomalies])

df = pd.DataFrame(X, columns=["Feature_1", "Feature_2"])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

lof = LocalOutlierFactor(
    n_neighbors=20,
    contamination=0.06
)

df["anomaly"] = lof.fit_predict(X_scaled)

normal = df[df["anomaly"] == 1]
outliers = df[df["anomaly"] == -1]

plt.figure(figsize=(8, 6))
plt.scatter(normal["Feature_1"], normal["Feature_2"],
            c="blue", s=30, label="Normal Points")
plt.scatter(outliers["Feature_1"], outliers["Feature_2"],
            c="red", s=60, marker="x", label="Anomalies")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Local Outlier Factor (LOF) ‚Äì Anomaly Detection")
plt.legend()
plt.show()
```

# 10- You are working as a data scientist for a power grid monitoring company. Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage. Explain your real-time data science workflow:
# ‚óè How would you detect anomalies in this streaming data (Isolation Forest / LOF / DBSCAN)?
# ‚óè Which time series model would you use for short-term forecasting (ARIMA / SARIMA / SARIMAX)?
# ‚óè How would you validate and monitor the performance over time?
# ‚óè How would this solution help business decisions or operations?



### Real-Time Data Science Workflow for Power Grid Monitoring

**Objective:**

* **Forecast short-term energy demand** (every 15 minutes)
* **Detect abnormal spikes/drops** in real-time consumption

The data includes **timestamp, region, weather variables, and energy usage**.


## 1Ô∏è‚É£ Anomaly Detection in Streaming Data

###  **Primary: Isolation Forest**

**Why:**

* Scales well for **high-frequency streaming data**
* Handles **multivariate inputs** (energy usage + weather)
* Effective for detecting **global anomalies** (sudden spikes/drops)

**How used in real time:**

* Train on historical ‚Äúnormal‚Äù data
* Score each incoming 15-minute record
* Flag points with high anomaly scores


###  **Secondary: Local Outlier Factor (LOF)**

**Why:**

* Detects **local anomalies**
* Useful for **region-wise abnormal behavior**

**How used:**

* Applied in **sliding windows** (e.g., last 1‚Äì2 hours per region)


###  **DBSCAN (Not ideal for real-time)**

* Computationally expensive
* Sensitive to parameter tuning
* Better for **offline spatial analysis**, not streaming

 **Final choice:**
 **Isolation Forest (system-wide)** + **LOF (regional/local)**


## 2Ô∏è‚É£ Time Series Model for Short-Term Forecasting

###  **Best Model: SARIMAX**

**Why SARIMAX:**

* Handles **seasonality** (daily & weekly cycles in power usage)
* Supports **exogenous variables**:

  * Temperature
  * Humidity
  * Weather conditions
  * Region indicators

**Comparison:**

* ARIMA ‚Üí no seasonality
* SARIMA ‚Üí seasonality, but no external factors
* **SARIMAX ‚Üí seasonality + weather effects (best fit)**


## 3Ô∏è‚É£ Validation & Continuous Monitoring

###  Forecast Validation

* Time-based train‚Äìtest split
* Metrics:

  * **MAE**
  * **RMSE**
  * **MAPE**

###  Continuous Monitoring

* Monitor:

  * Forecast residuals
  * Increase in anomaly frequency
* Detect **data drift**
* Retrain models:

  * Periodically (weekly)
  * Or when error crosses thresholds

###  Alerts

* Trigger alerts when:

  * Actual usage exceeds forecast confidence intervals
  * Sudden anomaly bursts are detected


## 4Ô∏è‚É£ Business & Operational Impact

###  Operational Benefits

* Early detection of:

  * Equipment faults
  * Power theft
  * Unexpected demand surges
* Prevent grid overloads and blackouts

###  Business Benefits

* Accurate demand planning
* Reduced operational cost
* Improved grid reliability
* Data-driven maintenance scheduling










```python
# Real-time style demo: Forecasting + Anomaly Detection

import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from statsmodels.tsa.statespace.sarimax import SARIMAX

np.random.seed(42)

date_range = pd.date_range("2024-01-01", periods=500, freq="15T")
energy = np.random.normal(200, 20, size=500)
temperature = np.random.normal(30, 5, size=500)

df = pd.DataFrame({
    "timestamp": date_range,
    "energy_usage": energy,
    "temperature": temperature
}).set_index("timestamp")

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

iso = IsolationForest(contamination=0.03, random_state=42)
df["anomaly"] = iso.fit_predict(X_scaled)

model = SARIMAX(
    df["energy_usage"],
    exog=df[["temperature"]],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 96)
)

results = model.fit(disp=False)

forecast = results.get_forecast(steps=8, exog=df[["temperature"]].iloc[-8:])
forecast_values = forecast.predicted_mean

print("Next 2-hour energy forecast:")
print(forecast_values)

print("\nNumber of anomalies detected:", (df["anomaly"] == -1).sum())
```
